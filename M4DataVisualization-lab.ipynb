## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
#Plot a histogram of ConvertedComp.
from matplotlib import pyplot as plt

QUERY = """
SELECT ConvertedComp
FROM master
"""
data = pd.read_sql_query(QUERY,conn)

# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(data, bins = 5)
 
# Show plot
plt.xlabel("ConvertedComp")
plt.ylabel("Count")
plt.title("Histogram of Converted Compensation")
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
#Plot a box plot of Age.
import matplotlib.pyplot as plt

QUERY = """
SELECT Age
FROM master
"""
data_age = pd.read_sql_query(QUERY,conn)
data_age

fig = plt.figure(figsize =(10, 7))
 
# Creating plot
data_age.plot(kind='box', title='Box Plot for Age') 

# Show plot

#plt.title("Boxplot for Age")
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
#Create a scatter plot of Age and WorkWeekHrs.
QUERY = """
SELECT Age,WorkWeekHrs
FROM master
"""
data1 = pd.read_sql_query(QUERY,conn)
# Load data into DataFrame 
df = pd.DataFrame(data = data1);
df['WorkWeekHrs']
# Draw a scatter plot 
df.plot.scatter(x = 'Age', y = 'WorkWeekHrs'); 
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
#Create a bubble plot of WorkWeekHrs and CodeRevHrs, use Age column as bubble size.
import seaborn as sns 
QUERY = """
SELECT WorkWeekHrs,CodeRevHrs,Age
FROM master
"""
data_bub = pd.read_sql_query(QUERY,conn)

# Load data into DataFrame 
df = pd.DataFrame(data = data_bub);

size = df['Age']
plt.figure(figsize=(8, 6)) 
sns.scatterplot(x="WorkWeekHrs", 
                y="CodeRevHrs", 
                data=df,
                sizes= size,
                alpha=0.5) 
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
#Create a pie chart of the top 5 databases that respondents wish to learn next year. 
#Label the pie chart with database names. Display percentages of each database on the pie chart.
QUERY = """
SELECT DatabaseDesireNextYear,COUNT(*) as count
FROM DatabaseDesireNextYear
group by DatabaseDesireNextYear
order by count desc
limit 5
"""
data_db = pd.read_sql_query(QUERY,conn)
# Load data into DataFrame 
df = pd.DataFrame(data = data_db);
label = df['DatabaseDesireNextYear']
df.plot(kind='pie',y = 'count',autopct='%1.0f%%',labels = label)


#df['DatabaseDesireNextYear']
```
